# Extended Username Validation

```python
import re

def validate_username(username):
    return re.fullmatch(r'^[A-Za-z][A-Za-z0-9_.]{4,19}$', username) is not None

print(validate_username('user_1')) # True
```

# API Key Authentication with a List

```python
from flask import request, jsonify

API_KEY = {'key1', 'key2', 'mysecret'}

def check_api_key(key=None):
    if key is None:
        key = request.headers.get('x-api-key')

    if key not in API_KEY:
        if key is None:
            return jsonify({'error': 'Unauthorized'}), 401
        else:
            return False

    return True

print(check_api_key("wrongkey"))  # False
print(check_api_key("key1"))      # True
```

# KeyError-Specific Error Handling

```python
def get_username(data):
    try:
        return data['username']
    except KeyError:
        return "Error: Missing username"
    except Exception:
        return "Error: Invalid data"    

# Modify so it catches only KeyError and returns "Error: Missing username";
# other exceptions should return "Error: Invalid data".
print(get_username({}))          # Error: Missing username
print(get_username({'username': 'Sam'}))
```

# Remove Multiple HTML Tags in Sanitisation

```python
import re
import html

def sanitise_comment(comment):
    comment = re.sub(r'<script.*?>.*?</script>', '', comment, flags=re.I|re.S)
    comment = re.sub(r'<img.*?>', '', comment, flags=re.I|re.S)
    return html.escape(comment)

# Extend this to also remove <img> tags before escaping.
print(sanitise_comment('<script>alert(1)</script>Hi!'))
```

# Age Validation with Feedback




```python
def validate_age(age):
    try:
        age_num = int(age)
    except ValueError:
        return "Age must be a number"   

    if 0 <= age_num <= 120:
        return "Valid"                   
    else:
        return "Age must be between 0 and 120"  
    
print(validate_age('abc'))   # Age must be a number
print(validate_age('25'))    # Valid
```
