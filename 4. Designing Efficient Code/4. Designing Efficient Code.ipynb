# Memory Management With Large Data

```python

def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()

for line in read_large_file('large_data.txt'):
    print(line)

# Task:
# - Rewrite this function to process the file line-by-line instead of loading everything at once.
# - Use a generator or iterate directly over the file object for memory efficiency.
```

# Session Management Optimisation 

```python
import time

sessions = {}

def create_session(user_id):
    sessions[user_id] = {'login_time': time.time(), 'data': {}}

def is_session_active(user_id, timeout=3600):
    if user_id in sessions:
        return (time.time() - sessions[user_id]['login_time']) < timeout
    return False

create_session('user123')

if is_session_active('user123'):
    print("Session is active.")
else:
    print("Session expired or does not exist.")
```

# Exception Handling Scope

```python
def divide_numbers(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print("Error: Cannot divide by zero.")
        return None
    else:
        print("Division successful.")
        return result
    finally:
        print("Execution of divide_numbers complete.")

divide_numbers(10, 2)
divide_numbers(10, 0)

# Task:
# - Modify to catch specific exceptions (e.g., ZeroDivisionError).
# - Return None or an appropriate value if an exception occurs.
# - Add a finally block that prints a message indicating function execution has finished.
```
