```python
def validate_age(age_input):
    

    try:
        age = int(age_input)
        if 0 <= age <= 120:
            return age
        else:
            raise ValueError("Age must be between 0 and 120.")
    except ValueError as e:
        print(f"Invalid input: {e}")
        return None

while True:
    user_input = input("please enter your age: ")
    age_input = validate_age(user_input)
    if age_input is not None:
        print(f"Your age is: {age_input}")
        break
    else:
        print("Please enter a valid age.")
```

```python
def validate_username(username_input):
    

    try:
        username = (username_input)
        if not username.isalnum():
            raise ValueError("Username is not alphanumeric")
        if not 3 <= len(username) <=15:
            raise ValueError("Username is not between 3 to 15 characters")
        else:
            return username
    except ValueError as e:
        print(f"Invalid username: {e}")
        return None

while True:
    user_input = input("Enter your username: ")
    username_input = validate_username(user_input)
    if username_input is not None:
        print(f"Welcome, {user_input}")
        break
    else:
        print("Please enter a valid username.")

```

```python
def validate_comments(comment_input):
    

    try:
        comments = (comment_input)
        if not comments.isalnum():
            raise ValueError("Comments are not alphanumeric")
        if len(comments) > 200:
            raise ValueError("Comments are longer than 200 characers")
        else:
            return comments
    except ValueError as e:
        print(f"Invalid comments: {e}")
        return None

while True:
    comment_input = input("Write your comment: ")
    comments_input = validate_comments(comment_input)
    if comments_input is not None:
        print(f"User comment: {comment_input}")
        break
    else:
        print("Please enter a valid comment.")

```

```python
def validate_num(number_input):
    

    try:
        number = int(number_input)
        if number < 0:
            raise ValueError("The number inserted is a negative integer")
        else:
            return number
    except ValueError as e:
        print(f"Invalid number: {e}")
        return None

while True:
    number_input = input("Enter a positive integer: ")
    num_input = validate_num(number_input)
    if num_input is not None:
        print(f"You entered: {number_input}")
        break
    else:
        print("Please enter a valid number.")

```
